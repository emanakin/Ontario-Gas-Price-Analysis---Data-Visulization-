Our dataset contains weekly retail prices of three grades of gasoline, diesel, auto propane and compressed natural gas in 10 Ontario markets. It is downloaded from https://data.ontario.ca/dataset/fuels-price-survey-information. Please plot the following visualizations. Your visualization should be complete.

```python
#IMPORTS
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

#DATASET
df = pd.read_csv('data/fueltypesall1991-2023.csv')
```

## Question 1

Compare the mean Regular Unleaded Gasoline price among different cities in 2022. Plot the corresponding bar chart.

```python
# Subset the dataframe to only include columns B - R and rows 0 - 1726
df_subset = df.iloc[:1726, 1:18]

# Calculate the mean of each column (city) in the subsetted dataframe
mean_prices = df_subset.mean()

# Plot a bar graph of the average prices
mean_prices.plot(kind='bar')

# Add a title and labels to the x-axis and y-axis
plt.title('Average Regular Unleaded Gasoline Price by City')
plt.xlabel('City')
plt.ylabel('Price ($/gallon)')

# Show the plot
plt.show()
```

## Question 2

Plot the histogram of the Regular Unleaded Gasoline price in Ottawa for 2022. Choose the appropriate parameters.

```python
# Subset the dataframe to only include the Ottawa data for 2022
ottawa_2022 = df_subset.iloc[1726-57:1726, 0]

# Plot a histogram of the Ottawa data for 2022
n, bins, patches = plt.hist(ottawa_2022, bins=15, range=(40, 240), edgecolor='black', linewidth=1.2)

# Add a title and labels to the x-axis and y-axis
plt.title('Histogram of Regular Unleaded Gasoline Price in Ottawa for 2022')
plt.xlabel('Price ($/gallon)')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

## Question 3

Plot the ECDF (normalized) of the Regular Unleaded Gasoline price in Ottawa for 2022.

```python
ottawa_2022_sorted = np.sort(ottawa_2022)

# Calculate the ECDF values
y = np.arange(1, len(ottawa_2022_sorted)+1) / len(ottawa_2022_sorted)

# Plot the ECDF
plt.step(ottawa_2022_sorted, y, where='post')

# Add a title and labels to the x-axis and y-axis
plt.title('ECDF of Regular Unleaded Gasoline Price in Ottawa for 2022')
plt.xlabel('Price ($/gallon)')
plt.ylabel('Normalized Cumulative Distribution')

# Show the plot
plt.show()
```
